### Pregunta 1

Para encontrar la función de consumo primero tenemos que entender lo que es el ingreso. El ingreso es la suma de los salarios de los trabajadores y el stock de capital, es decir, todo el ingreso del país, que se denomina Ingreso Agregado (Y). Una parte de este ingreso se entrega al Estado como tributos (T), en base a la tasa de impuestos (t). Por lo tanto, tenemos la siguiente función de Tributación total:

$$ T = tY $$

A partir de esto, podemos encontrar el Ingreso Disponible (Yd), que corresponde al ingreso que realmente queda en el bolsillo de la gente luego de la tributación:

$$ Y^d = Y-T $$

Esta ecuación también se puede mostrar de la siguiente manera:

$$ Y^d = Y(1 - t) $$

Entendiendo lo que representa el Ingreso Disponible, ahora podemos generar la función de consumo, que suma el consumo autónomo (Co), es decir, el gasto necesario de las familias para su supervivencia, con la variación en los gastos del ingreso disponible de las familias, que se conoce como la propensión marginal a consumir (b), que tiene un valor del 0 al 1, multiplicado por el ingreso disponible.

$$ C = C_0 + bY^d $$

El consumo autónomo es una constante, pero luego encontramos la relación entre la propensión marginal a consumir y el Ingreso Disponible. Si cambia "b", entonces el consumo del ingreso disponible cambiará, ya que multiplica a este valor. Si la propensión es más cercana a 1, entonces las personas gastarán más de su ingreso disponible para su consumo. Por lo tanto, la función de consumo depende del consumo autónomo y de la relación entre el Ingreso disponible y la PMgC.





```python
# Parámetros

Yd_size = 70 

Co = 10
b = 0.5 # b > m

Yd = np.arange(Yd_size)

def C_K(Co, b, Yd):
    C_K = Co + (b*Yd)
    return C_K

C_IS_K = C_K(Co, b, Yd)
```

```python
# Gráfico

y_max = np.max(C_IS_K)
fig, ax = plt.subplots(figsize=(10, 7))

ax.plot(Yd, C_IS_K, label = "C=Co+bYd", color = "#3D59AB") 

plt.text(57, 50, '$PMgC = {∆C}/{∆Y^d} = b = 0.5$', fontsize = 12, color = 'black')

ax.set(title="Función de demanda de Consumo", xlabel= r'Yd', ylabel= r'C')
ax.legend() 

custom_xlim = (0, 80)
custom_ylim = (0, 80)

plt.setp(ax, xlim=custom_xlim, ylim=custom_ylim)

plt.show()
```

### Pregunta 2

La función de Inversión tiene 3 componentes, que representan los gastos que aumentan el stock de capital de la economía, y por lo tanto la producción:

Inversión autónoma (Io): represente el costo de reemplazar y mantener funcionando lo que se va desgastando o malogrando. Es la variación de las inversiones fijas, constantes.

h: es la sensibilidad de los inversionistas ante la tasa de interés. Va de 0 a 1. 

r: es la tasa de interés real o activa. Va del 0 al 1. 

Por lo tanto, la fórumla de la Inversión sería la inversión autónoma menos la multiplicación de la sensibilidad para invertir con la tasa de interés real:

$$ I = I_0 - hr $$

Esto implica que si sube la tasa de interés, los inversionistas van a preferir invertir menos. Por lo tanto, la inversión en un país y la tasa de interés tienen relación inversa.

```python
# Parámetros

r_size = 100 


Io = 80
h = 0.7
r = 0.9

r = np.arange(r_size)

def I_K(Io, h, r):
    I_K = Io - (h)*(r)
    return I_K

I_IS_K = I_K(Io, h, r)
```

```python
# Gráfico

y_max = np.max(I_IS_K)
fig, ax = plt.subplots(figsize=(10, 6))

ax.plot(r, I_IS_K, label = "I=Io-hr", color = "#3D59AB")

ax.yaxis.set_major_locator(plt.NullLocator())   
ax.xaxis.set_major_locator(plt.NullLocator())

plt.text(97, 20, '$I=I_0-hr$', fontsize = 12, color = 'black')

ax.set(title="Función de demanda de Inversión", xlabel= r'r', ylabel= r'I')
ax.legend() 

custom_xlim = (0, 120)
custom_ylim = (-10, 120)

plt.setp(ax, xlim=custom_xlim, ylim=custom_ylim)

plt.show()
```

### Pregunta 4

Vamos a encontrar el Ye en base a las ecuaciones del modelo Ingreso-Gasto

Para llegar al equilibrio, debemos igualar DA = Y, es decir, la demanda agregada debe ser igual el ingreso agregado.

Primero tenemos las ecuaciones del gasto:

$$ C = C_0 + bY^d $$
$$ I = I_0 - hr $$
$$ G = G_0 $$
$$ X = X_0 $$
$$ M = mY^d $$

Si sumamos todas estas, encontramos la función de Demanda Agregada:

$$ DA = C + I + G + X - M $$

Ahora tomamos la función de Ingreso disponible:

$$ Y^d = Y(1 - t) $$

y la reemplazamos en la función de la demanda agregada:

$$ DA = C_0 + I_0 + G_0 + X_0 - hr + Y(b - m)(1 - t) $$

Esta función se puede escribir de la siguiente manera, con dos partes alpha:

$$ DA = α_0 + α_1Y $$

α_0 sería el intercepto, que son todos los gastos autónomos, y α_1 sería la pendiente.

Recordemos que el equilibrio es DA = Y, por lo que podemos decir que:

$$ Y = C_0 + bY^d + I_0 -hr + G_0 + X_0 - mY^d $$

$$ Y^e = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

Donde "1/ 1-b-m 1-t" representa al multiplicador keynesiano "k". Esta sería la ecuación de equilibrio del modelo Ingreso-Gasto keynesiano.

```python
# Parámetros

Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8 # b > m
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

# Ecuación de la curva del ingreso de equilibrio

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
a = 2.5 

def L_45(a, Y):
    L_45 = a*Y
    return L_45

L_45 = L_45(a, Y)

y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 10))

ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #Demanda agregada
ax.plot(Y, L_45, color = "#404040") #Línea de 45º



plt.axvline(x=70.5,  ymin= 0, ymax= 0.69, linestyle = ":", color = "grey")
plt.axhline(y=176, xmin= 0, xmax= 0.7, linestyle = ":", color = "grey")
plt.axhline(y=146, xmin= 0, xmax= 0.9, linestyle = ":", color = "grey")

plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(0, 152, '$α_o$', fontsize = 15, color = 'black')

plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')

plt.text(86, 200, '$DA = α_0 + α_1Y$', fontsize = 12, color = 'black')
plt.text(97, 170, '$α_1 = (b-m)(1-t)$', fontsize = 12, color = 'black')
plt.text(72,120, '$α_0 = (C_0 + I_0 + G_0 + X_0 - hr)$', fontsize = 12, color = 'black')

ax.annotate("", xy=(85, 143), xytext=(78, 130),
            arrowprops=dict(arrowstyle="->"))

ax.annotate("", xy=(80, 180), xytext=(90, 200),
            arrowprops=dict(arrowstyle="->"))


ax.set(title="El ingreso de Equilibrio a Corto Plazo", xlabel= r'Y', ylabel= r'DA')
ax.legend()

plt.show()
```

Aquí el punto de equilibrio tiene un intercepto de 147.63, mientras que la pendiente equivale a 0.42 (Y).

### Pregunta 5

#### Política Fiscal Expansiva con aumento del Gasto de Gobierno:

Intuitivo: $$ ↑Go → ↑DA → DA > Y → ↑Y $$

Si aumenta el gasto del gobierno, al ser un valor positivo de la ecuación de DA, entonces aumenta la demanda agregada. Entonces, si la DA aumenta, va a ser mayor que el Ingreso Agregado. Por consecuencia, el Ingreso debe de aumentar para regresar al equilibrio.

Matemático: $∆G_0 > 0  →  ¿∆Y?$ Debemos identificar cómo va a variar el Ingreso para regresar al equilibrio si aumenta el Gasto autónomo.

$$ Y = \frac{1}{1 - (b - m)(1 - t)} (C_0 + I_0 + G_0 + X_0 - hr) $$

Pero, si no ha habido cambios en $C_0$, $I_0$, $X_0$, $h$ ni $r$, entonces la variación de Y solo se dará en términos del gasto autónomo: 

$$∆C_0 = ∆I_0 = ∆X_0 = ∆h = ∆r = 0$$

$$ ∆Y = k (∆G_0) $$

Sabiendo que $∆G_0 > 0 $, es decir, que aumento el gasto, y que $k > 0$, la multiplicación de un número positivo con otro positivo dará positivo:

$$ ∆Y = (+)(+) $$
$$ ∆Y > 0 $$

Por lo tanto, el ingreso aumentará como consecuencia del aumento del gasto de gobierno.

```python
#Gráfico: vamos a graficar la curva de equilibrio original, y luego 
#vamos a graficar la nueva curva, con un aumento del Gasto de gobierno.

#--------------------------------------------------
# Curva de ingreso de equilibrio ORIGINAL

    # Parámetros
Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9

Y = np.arange(Y_size)

    # Ecuación 
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)


#--------------------------------------------------
# NUEVA curva de ingreso de equilibrio con aumento del Gasto de gobierno

Go = 80

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

#### Política Fiscal expansiva con una reducción de la Tasa de Tributación:

Intuitivo: $$ t↓ → Co↑ → DA↑ → DA > Y → Y↑ $$
            $$ t↓ → M↑ → DA↓ → DA < Y → Y↓ $$
            
Al bajar la tasa impositiva, el consumo aumenta, ya que las personas mantienen más dinero disponible para su consumo. Por lo tanto, aumenta la demanda agregada, y por consecuencia, requiere que aumente el Ingreso Agregado para volver al equilibrio.

Asimismo, una disminución de la tasa impositiva genera que aumenten las importaciones. Esto reduce la demanda agregada, y genera un déficit de demanda, que solo regresará al equilibrio con una disminución del Ingreso.

Sin embargo, debido a que "b" es mayor que "m", concluyo que el impacto del consumo siempre va a ser mayor que el impacto de las importaciones. Así disminuya Y por las importaciones, el aumento de Y debido al consumo hará que el diferencial de Y sea finalmente positivo.

Matemática: $∆t < 0  →  ¿∆Y?$

Tenemos el siguiente diferencial de la tasa de tributación:

$$ \frac{∆Y}{∆t} = \frac{(m-b)(Co + Go + Io + Xo - hr)}{(1-(1-t)(b-m)+1)^2} $$

Sin embargo, sabemos que "b" es mayor que "m". Por lo que saldrá un número negativo al hacer m-b. Por lo tanto, la primera parte de nuestro ecuación es negativa (-).

Además, los componentes autónomos no cambian, lo que nos da un diferencial de 0. Por lo tanto, el numerador será negativo.

En el denominador será positivo (+) en cambio, porque se eleva al cuadrado, lo que siempre da un resultado positivo.

Por lo tanto, tenemos la siguiente división:

$$ \frac{∆Y}{∆t} = \frac{(-)}{(+)} $$

Pero sabemos que el diferencial de la tasa impositiva es menor que 0, por lo que la ecuación se vuelve esto:

$$ \frac{∆Y}{(-)} = \frac{(-)}{(+)} $$

$$ ∆Y = \frac{(-)(-)}{(+)} $$

Por lo tanto, el resultado final del diferencial será un número positivo entre otro positivo, por lo que el diferencial de Y es positivo.

$$ ∆Y > 0 $$

```python
#Gráfico de la reducción de la tasa de tributación.

#--------------------------------------------------
# Curva de ingreso de equilibrio ORIGINAL

    # Parámetros
Y_size = 100 

Co = 35
Io = 40
Go = 70
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3 #tasa de tributación
r = 0.9

Y = np.arange(Y_size)

    # Ecuación 
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_IS_K = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)


#--------------------------------------------------
# NUEVA curva de ingreso de equilibrio

t = 0.01

# Generar la ecuación con el nuevo parámetros
def DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + ((b - m)*(1 - t)*Y)
    return DA_K

DA_t = DA_K(Co, Io, Go, Xo, h, r, b, m, t, Y)
```

```python
# Gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))


# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #curva ORIGINAL
ax.plot(Y, DA_t, label = "DA_t", color = "#EE7600", linestyle = 'dashed') #NUEVA curva
ax.plot(Y, L_45, color = "#404040") #línea de 45º

# Lineas punteadas
plt.axvline(x = 70.5, ymin= 0, ymax = 0.69, linestyle = ":", color = "grey")
plt.axhline(y = 176, xmin= 0, xmax = 0.7, linestyle = ":", color = "grey")
plt.axvline(x = 77,  ymin= 0, ymax = 0.75, linestyle = ":", color = "grey")
plt.axhline(y = 192, xmin= 0, xmax = 0.75, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 180, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(0, 200, '$DA_t$', fontsize = 11.5, color = '#EE7600')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
plt.text(72, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(80, 0, '$Y_t$', fontsize = 12, color = '#EE7600')
plt.text(72, 45, '$→$', fontsize = 18, color = 'grey')
plt.text(20, 180, '$↑$', fontsize = 18, color = 'grey')

# Título y leyenda
ax.set(title = "Reducción de la Tasa de Tributación", xlabel = r'Y', ylabel = r'DA')
ax.legend()

plt.show()
```

#### Equilibrio con regla contracíclica

Primero tenemos las ecuaciones del gasto:

$$ C = C_0 + bY^d $$
$$ I = I_0 - hr $$
$$ G = G_0 - gY $$
$$ X = X_0 $$
$$ M = mY^d $$

Pero ahora el gasto del gobierno tiene un nuevo componente "-gY".

Si sumamos todas estas, encontramos la función de Demanda Agregada:

$$ DA = C + I + G + X - M $$

Ahora tomamos la función de Ingreso disponible:

$$ Y^d = Y(1 - t) $$

y la reemplazamos en la función de la demanda agregada:

$$ DA = C_0 + I_0 + G_0 + X_0 - hr + Y((b - m)(1 - t)-g) $$

Esta función se puede escribir de la siguiente manera, con dos partes alpha:

$$ DA = α_0 + α_0Y $$

α_0 sería el intercepto y α_1 sería la pendiente.

Recordemos que el equilibrio es DA = Y, por lo que podemos decir que:

$$ Y = C_0 + bY^d + I_0 -hr + G_0 - gY + X_0 - mY^d $$

$$ Y^e = \frac{1}{1 - (b - m)(1 - t) + g} (C_0 + I_0 + G_0 + X_0 - hr) $$

Lo que cambia es que tenemos a "g" en el denominador del multiplicador keynesiano. Esta sería la ecuación de equilibrio del modelo Ingreso-Gasto keynesiano.

#### Política Fiscal Expansiva con aumento del Gasto de Gobierno:

Intuitivo: $$ ↑Go → ↑DA → DA > Y → ↑Y $$

Si aumenta el gasto del gobierno esto genera que aumente la demanda agregada. Entonces, si la DA aumenta, va a ser mayor que el Ingreso Agregado. Por consecuencia, el Ingreso debe de aumentar para regresar al equilibrio. Con la adición del parámetro "g", la pendiente de la curva va a disminuír, pero igual obliga al ingreso agregado a aumentar para regresar al equilibrio.

Matemático: $∆G_0 > 0  →  ¿∆Y?$ Debemos identificar cómo va a variar el Ingreso para regresar al equilibrio si aumenta el Gasto autónomo.

$$ Y = \frac{1}{1 - (b - m)(1 - t) + g} (C_0 + I_0 + G_0 + X_0 - hr) $$

Ahora que tenemos a "g" en la ecuación, su efecto es que aumenta la magnitud del denominador, por lo que el tamaño de la pendiente se reduce.

Igual que antes, no ha habido cambios en $C_0$, $I_0$, $X_0$, $h$ ni $r$, entonces la variación de Y solo se dará en términos del gasto autónomo: 

$$∆C_0 = ∆I_0 = ∆X_0 = ∆h = ∆r = 0$$

$$ ∆Y = k (∆G_0) $$

Sabiendo que $∆G_0 > 0 $  y que $k > 0$, la multiplicación de un número positivo con otro positivo dará positivo:

$$ ∆Y = (+)(+) $$
$$ ∆Y > 0 $$

Por lo tanto, el ingreso aumentará como consecuencia del aumento del gasto de gobierno.

```python
#Gráfico: vamos a graficar la curva de equilibrio original, y luego 
#vamos a graficar la nueva curva, con el aumento del gasto de gobierno.

#--------------------------------------------------
# Curva de ingreso de equilibrio ORIGINAL

    # Parámetros
Y_size = 100 

Co = 35
Io = 40
Go = 50
Xo = 2
h = 0.7
b = 0.8
m = 0.2
t = 0.3
r = 0.9
g = 0.1

Y = np.arange(Y_size)

    # Ecuación 
def DA_IS_K(Co, Io, Go, Xo, h, r, b, m, t, g, Y):
    DA_IS_K = (Co + Io + Go + Xo - h*r) + (((b - m)*(1 - t)-g)*Y)
    return DA_IS_K

DA_IS_K = DA_IS_K(Co, Io, Go, Xo, h, r, b, m, t, g, Y)


#--------------------------------------------------
# NUEVA curva de ingreso de equilibrio con aumento del Gasto de gobierno

Go = 80

def DA_K(Co, Io, Go, Xo, h, r, b, m, t, g, Y):
    DA_K = (Co + Io + Go + Xo - h*r) + (((b - m)*(1 - t)-g)*Y)
    return DA_K

DA_G = DA_K(Co, Io, Go, Xo, h, r, b, m, t, g, Y)
```

```python
# Gráfico
y_max = np.max(DA_IS_K)
fig, ax = plt.subplots(figsize=(10, 8))


# Curvas a graficar
ax.plot(Y, DA_IS_K, label = "DA", color = "#3D59AB") #curva ORIGINAL
ax.plot(Y, DA_G, label = "DA_G", color = "#EE7600", linestyle = 'dashed') #NUEVA curva
ax.plot(Y, L_45, color = "#404040") #línea de 45º

# Lineas punteadas
plt.axvline(x = 72, ymin= 0, ymax = 0.71, linestyle = ":", color = "grey")
plt.axhline(y = 180, xmin= 0, xmax = 0.71, linestyle = ":", color = "grey")
plt.axvline(x = 58,  ymin= 0, ymax = 0.57, linestyle = ":", color = "grey")
plt.axhline(y = 145, xmin= 0, xmax = 0.57, linestyle = ":", color = "grey")

# Texto agregado
plt.text(0, 133, '$DA^e$', fontsize = 11.5, color = 'black')
plt.text(0, 170, '$DA_G$', fontsize = 11.5, color = '#EE7600')
plt.text(6, 4, '$45°$', fontsize = 11.5, color = 'black')
plt.text(2.5, -3, '$◝$', fontsize = 30, color = '#404040')
plt.text(70, 0, '$Y_G$', fontsize = 12, color = '#EE7600')
plt.text(57, 0, '$Y^e$', fontsize = 12, color = 'black')
plt.text(65, 45, '$→$', fontsize = 18, color = 'grey')
plt.text(20, 140, '$↑$', fontsize = 18, color = 'grey')

# Título y leyenda
ax.set(title = "Aumento del Gasto del Gobierno $(G_0)$", xlabel = r'Y', ylabel = r'DA')
ax.legend()

plt.show()
```

El parámetro "g" aumenta el tamaño del denominador del multiplicador keynesiano, ya que se lo suma a los valores del denominador. Por lo tanto, si aumenta la magnitud del denominador, entonces se reducirá el tamaño de "k", por lo que la pendiente de la recta de DA también se reducirá.

A diferencia del modelo de la pregunta 5, el valor de "k" es menor, ya que el denominador ha aumentado por la adición de "g". Por lo tanto, si "k" es menor, el impacto de un aumento del gasto de gobierno se reducirá, a diferencia de cuando no existía el parámetro "g". El tamaño del efecto de la política fiscal se reducirá, ya que el parámetro "g" reduce la pendiente. Por lo tanto, el efecto de la política fiscal de aumento de gasto de goberno en el modelo 6 es menor que el tamaño del efecto en el modelo 5.
